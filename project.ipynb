```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df=pd.read_csv("Top_10000_Movies.csv")
```

```python
df.describe().T
```

```python
df.drop(["Unnamed: 0","id"], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["release_date"] =  pd.to_datetime(df["release_date"])
```

```python
df["release_date"].dtype
# corr=df.corr()
# sns.heatmap(corr,annot=True,cmap='Reds')
```

```python
plt.figure(figsize=(6,5))
color=np.array(['red','green','blue','yellow','cyan'])
df["original_language"].value_counts().head(5).plot(kind='bar',color=color)
plt.xticks(rotation=0)
plt.title("Most Used 5 Languages", size=15)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.distplot(df["runtime"])
plt.title("Runtime in Minutes", size=15)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=df["vote_count"], y=df["vote_average"])
plt.title("Vote Count vs Vote Average", size=15)
plt.annotate('Most Rated With Max Votes',(30300,8.3),(25000,4),arrowprops={'arrowstyle':'<->','color':'red'},size=20,color='r')
plt.show()
```

```python
top_15_movies = df[df["genre"] != "[]"].sort_values(by="vote_average", ascending=False).head(15)
print(top_15_movies)
plt.figure(figsize=(10,10))
sns.countplot(top_15_movies["genre"])
plt.title("Genres of Top-Rated 15 Movies", size=15)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5,4))
Count=df["release_date"].dt.month.value_counts().sort_index()
X=Count.index
plt.plot(X,Count,ls='--',marker='*')
plt.title("NO. of movies release in months ", size=15)
plt.ylabel("NO. of movies release")
plt.xlim(0,13,3)
plt.xlabel("NO. of month")
plt.show()
```

```python

plt.figure(figsize=(6,3))
sns.boxplot(df["vote_average"])
plt.title("Rating of all movies", size=15)
plt.xlim(0,10)
plt.xlabel("Rating")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,8))
highest_revenued = df[df["genre"]!="[]" ].sort_values(by="revenue", ascending=False).head(30)
count=highest_revenued["genre"].value_counts()
lang=count.index
lang=lang[0:5]
count.index = np.arange(0,22)
count=count.drop(np.arange(5,22))
plt.pie(count,labels=lang, startangle=90, autopct='%1.1f%%',counterclock=False)
plt.title("Highest Revenued Genres",size=15)
plt.show()
```
